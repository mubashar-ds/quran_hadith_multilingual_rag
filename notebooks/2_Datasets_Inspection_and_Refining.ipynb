# **1 - Hadith Dataset Inspection**

```python
import json
import pandas as pd

DATA_PATH = "/content/4_Sahih_Bukhari_sorted_fixed_final.json"

def load_dataset(path):

    if path.endswith(".jsonl"):
        data = [json.loads(line) for line in open(path, "r", encoding="utf-8")]
    elif path.endswith(".json"):
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
    else:
        raise ValueError("\nUnsupported file format. Use .json or .jsonl")
    return data

data = load_dataset(DATA_PATH)
df = pd.DataFrame(data)

print(f"Total records loaded: {len(df)}\n")
df.head()
```

**1.1 - Hadith Dataset Refining**

```python
import json
import pandas as pd

DATA_PATH = "/content/4_Sahih_Bukhari_sorted_fixed_final.json"
OUTPUT_PATH = "/content/5_Sahih_Bukhari_sorted_fixed_final_1.json"

REMOVE_IDS = [
    538, 658, 836, 854, 902, 1256, 1266, 1332, 1373, 1656,
    2072, 2136, 2186, 2355, 3202, 4096, 5718, 7241, 7407
]

with open(DATA_PATH, 'r', encoding = 'utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)
print("Original total:", len(df))

df = df[~df["bukhari_id"].isin(REMOVE_IDS)].reset_index(drop = True)
print("\nAfter deletion:", len(df))

df["bukhari_id"] = range(1, len(df) + 1)

df.to_json(OUTPUT_PATH, orient="records", indent = 2, force_ascii = False)
print("\nSaved cleaned dataset to:", OUTPUT_PATH)
```

```python
reference_ids = list(range(1, 7564))
ref_set = set(reference_ids)

your_set = set(df["hadith_id"].tolist())

missing = sorted(ref_set - your_set)
extra = sorted(your_set - ref_set)

print("Missing in your dataset:", missing[:50])
print("\nExtra in your dataset:", extra[:50])
print("\nTotal missing:", len(missing))
print("\nTotal extra:", len(extra))
```

# **2 - Quran Dataset Reformatting**

```python
import json
import pandas as pd

INPUT_FILE = "/content/2_Quran_merged_juz.json"
OUTPUT_FILE = "/content/3_Al_Quran_Dataset_FINAL.json"

def load_json_auto(path):

    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
            if isinstance(data, list):
                return data
            else:
                raise Exception("JSON file is not a list of objects.")
    except:

        data = []
        with open(path, "r", encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if line:
                    data.append(json.loads(line))
        return data

print("Loading data...")
data = load_json_auto(INPUT_FILE)
print(f"\nLoaded {len(data)} records.")

def clean_quran_record(rec):

    return {
        "quran_id": rec.get("quran_id"),
        "juz": rec.get("juz"),
        "surah_id": rec.get("surah_id"),
        "ayah_id": rec.get("ayah_id"),
        "source": "Quran",

        "transliteration": rec.get("transliteration"),
        "surah_name_ar": rec.get("surah_name_ar"),
        "surah_name_ur": rec.get("surah_name_ur"),
        "surah_name_en": rec.get("surah_name_en"),
        "surah_type": rec.get("surah_type"),

        "text_ar": rec.get("text_ar"),
        "text_ur": rec.get("text_ur"),
        "text_en": rec.get("text_en")
    }

print("\nCleaning the records...")
cleaned = [clean_quran_record(r) for r in data]

with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(cleaned, f, ensure_ascii=False, indent=4)

print("\nCleaning complete!")
print(f"\nSaved reformatted dataset -> {OUTPUT_FILE}")
```

# **3 - Hadith Dataset Reformatting**

```python
import json
import pandas as pd

INPUT_FILE = "/content/5_Sahih_Bukhari_sorted_fixed_final_1.json"
OUTPUT_FILE = "/content/6_Al_Hadith_Dataset_FINAL.json"

def load_json_auto(path):

    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
            if isinstance(data, list):
                return data
            else:
                raise Exception("JSON file is not a list.")
    except:

        data = []
        with open(path, "r", encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if line:
                    data.append(json.loads(line))
        return data

print("Loading Hadith dataset...")
data = load_json_auto(INPUT_FILE)
print(f"\nLoaded {len(data)} records.")

def clean_hadith_record(rec):

    return {

        # classification..
        "hadith_id": rec.get("hadith_id"),
        "source": "Hadith",
        "book": rec.get("book"),
        "book_id": rec.get("book_id"),
        "chapter_id": rec.get("chapter_id"),

        # chapter names...
        "chapter_name_ar": rec.get("chapter_name_ar"),
        "chapter_name_ur": rec.get("chapter_name_ur"),
        "chapter_name_en": rec.get("chapter_name_en"),

        # texts
        "arabic_text": rec.get("arabic_text"),
        "urdu_text": rec.get("urdu_text"),
        "english_text": rec.get("english_text"),

        # metadata,.
        "narrator": rec.get("narrator"),
        "status": rec.get("status")
    }

print("\nCleaning records...")
cleaned = [clean_hadith_record(r) for r in data]

with open(OUTPUT_FILE, 'w', encoding = 'utf-8') as f:
    json.dump(cleaned, f, ensure_ascii = False, indent=4)

print("\nCleaning completed!")
print(f"\nSaved reformatted dataset -> {OUTPUT_FILE}")
```
